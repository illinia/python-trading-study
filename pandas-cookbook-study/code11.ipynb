```python
years = 2016, 2017, 2018
```

```python
stock_tables = [pd.read_csv(f'data/stocks_{year}.csv', index_col='Symbol') for year in years]
```

```python
stocks_2016, stocks_2017, stocks_2018 = stock_tables
```

```python
stock_tables
```

```python
stocks_2017
```

```python
stocks_2018
```

```python
pd.concat(stock_tables, keys=[2016, 2017, 2018])
```

```python
pd.concat(dict(zip(years, stock_tables)), axis='columns')
```

```python
list(zip(years, stock_tables))
```

```python
stocks_2016.join(
    stocks_2017,
    lsuffix='_2016',
    rsuffix='_2017',
    how='outer'
)
```

```python
other = [stocks_2017.add_suffix('_2017'),
         stocks_2018.add_suffix('_2018')]
other
```

```python
stocks_2016.add_suffix('_2016').join(other, how='outer')
```

```python
stock_join = stocks_2016.add_suffix('_2016').join(other, how='outer')
```

```python
stock_concat = (
    pd.concat(
        dict(zip(years, stock_tables)), axis='columns'
    )
    .swaplevel(axis=1)
    .pipe(lambda df_: df_.set_axis(df_.columns.to_flat_index(), axis=1))
    .rename(lambda label: '_'.join([str(x) for x in label]), axis=1)
)
stock_concat
```

```python
stock_join.equals(stock_concat)
```

```python
stocks_2016.merge(stocks_2017,
                  left_index=True,
                  right_index=True
                  )
```

```python
stock_merge = (
    stocks_2016
    .merge(
        stocks_2017,
        left_index=True,
        right_index=True,
        how='outer',
        suffixes=('_2016', '_2017')
    )
    .merge(
        stocks_2018.add_suffix('_2018'),
        left_index=True,
        right_index=True,
        how='outer'
    )
)
stock_merge
```

```python
stock_concat.sort_index().equals(stock_merge)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///data/chinook.db')
```

```python
tracks = pd.read_sql_table('tracks', engine)
tracks
```

```python
(
    pd.read_sql_table('genres', engine)
    .merge(tracks[['GenreId', 'Milliseconds']],
           on='GenreId',
           how='left')
    .drop('GenreId', axis='columns')
)
```

```python
(
    pd.read_sql_table('genres', engine)
    .merge(tracks[['GenreId', 'Milliseconds']],
           on='GenreId',
           how='left')
    .drop('GenreId', axis='columns')
    .groupby('Name')
    ['Milliseconds']
    .mean()
    .pipe(lambda s_: pd.to_timedelta(s_, unit='ms').rename('Length'))
    .dt.floor('s')
    .sort_values()
)
```

```python
cust = pd.read_sql_table('customers', engine, columns=['CustomerId', 'FirstName', 'LastName'])
```

```python
invoice = pd.read_sql_table('invoices', engine, columns=['InvoiceId', 'CustomerId'])
```

```python
invoice_items = pd.read_sql_table('invoice_items', engine, columns=['InvoiceId', 'UnitPrice', 'Quantity'])
```

```python
(
    cust
    .merge(invoice, on='CustomerId')
    .merge(invoice_items, on='InvoiceId')
)
```

```python
(
    cust
    .merge(invoice, on='CustomerId')
    .merge(invoice_items, on='InvoiceId')
    .assign(Total=lambda df_: df_.Quantity * df_.UnitPrice)
    .groupby(['CustomerId', 'FirstName', 'LastName'])
    ['Total']
    .sum()
    .sort_values(ascending=False)
)
```
